```python
df_tabla_final = pd.read_excel("tabla_final.xlsx", sheet_name="Hoja2")
df_tabla_final
```

```python
subcarpetas = ["MEDICIONES 2023 ADICIONALES", "MEDICIONES 2023 CONTRATO", "MEDICIONES 2023 PRIMERA MUESTRA"]
dict_meses = {"01": "Enero",
              "02": "Febrero",
              "03": "Marzo",
              "04": "Abril",
              "05": "Mayo",
              "06": "Junio",
              "07": "Julio",
              "08": "Agosto",
              "09": "Septiembre",
              "10": "Octubre",
              "11": "Noviembre",
              "12": "Diciembre"}

df_tabla_final["CODIGOCLIE"] = df_tabla_final["CODIGOCLIE"].astype("string")

for index, row in df_tabla_final.iterrows():

    mult_potencias = row["Mult. de potencias"]

    for subcarpeta in subcarpetas:
        archivos = os.scandir(subcarpeta)
        for archivo in archivos:
            nombre_archivo = archivo.name
            cue = nombre_archivo.split("-")[1]

            if cue.startswith("0"):
                cue = cue[1:]

            if cue in row["CODIGOCLIE"]:
                #print("MATCH")

                df_archivo_telem_tab = pd.DataFrame([1,2,3,4], columns=["Prueba"])
                df_archivo_telem_pc = pd.DataFrame([1,2,3,4], columns=["Prueba"])
                df_archivo_telem_c = pd.DataFrame([1,2,3,4], columns=["Prueba"])

                mes = dict_meses[nombre_archivo.split("-")[2][4:6]]

                path_archivo = f"{subcarpeta}/{nombre_archivo}"
                print(path_archivo)

                try:
                    #print("Leyendo con tab")
                    df_archivo_telem_tab = pd.read_csv(path_archivo,
                                                                    sep="\t",
                                                                    #decimal=",",
                                                                    skiprows=11,
                                                                    #na_values="N/D",
                                                                    encoding="utf-16",
                                                                    #on_bad_lines="skip",
                                                                    encoding_errors="ignore" 
                                                                    ) 
                except:
                    pass

                try:
                    #print("Leyendo con ;")
                    df_archivo_telem_pc = pd.read_csv(path_archivo,
                                                                    sep=";",
                                                                    #decimal=",",
                                                                    skiprows=11,
                                                                    #na_values="N/D",
                                                                    #encoding="utf-16",
                                                                    #on_bad_lines="skip",
                                                                    encoding_errors="ignore" 
                                                                    )

                except:
                    pass

                try:
                    #print("Leyendo con ,")
                    df_archivo_telem_c = pd.read_csv(path_archivo,
                                                                    sep=",",
                                                                    #decimal=",",
                                                                    skiprows=11,
                                                                    #na_values="N/D",
                                                                    #encoding="utf-16",
                                                                    #on_bad_lines="skip",
                                                                    encoding_errors="ignore" 
                                                                    )

                except:
                    pass

                if len(df_archivo_telem_tab.columns) > 1:
                    df_archivo_telem = df_archivo_telem_tab.copy()
                elif len(df_archivo_telem_pc.columns) > 1:
                    df_archivo_telem = df_archivo_telem_pc.copy()
                else:
                    df_archivo_telem = df_archivo_telem_c.copy()
                
                #print(df_archivo_telem.columns)
                #print(df_archivo_telem)
                max_demanda = df_archivo_telem.loc[df_archivo_telem.iloc[:,3].idxmax()]

                demanda_activa_max = max_demanda.iloc[3]*mult_potencias
                demanda_reactiva_max = max_demanda["Demanda reactiva DEL"]*mult_potencias
                fecha_valor_max = max_demanda["Fecha"]
                print(f"{demanda_activa_max},   {fecha_valor_max}")

                df_tabla_final.loc[index, f"Demanda_activa_max_{mes}"] = demanda_activa_max
                df_tabla_final.loc[index, f"Demanda_reactiva_max_{mes}"] = demanda_reactiva_max
                df_tabla_final.loc[index, f"Fecha_demanda_max_{mes}"] = fecha_valor_max

                #print(df_archivo_telem)
            else:
                pass
```

```python
df_tabla_final.to_excel("tabla_final_llenada.xlsx", index=False)
```
